# Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the shopping dataset

```python
dataset = pd.read_csv('shopping_data.csv')
```

```python
X = dataset.iloc[:,[3,4]].values
```

# Using the elbow method to find the optimal number of cluster

```python
# wcss-within cluster sum of squares, 11 becasue we want from 1 to 10,init-randome cluster selection, we use kmean++,
#max iter-default
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans =KMeans(n_clusters =i, init = 'k-means++', max_iter =300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Applying K-means to the dataset

```python
kmeans = KMeans(n_clusters =5 ,init = 'k-means++',max_iter =300, n_init =10, random_state =0)
y_kmeans = kmeans.fit_predict(X)
#Below we get the assignment of clusters from 1 to 4
print(y_kmeans)
```

# Visualising the cluster

```python
# first coloumn of data x(, 0), 2nd coloumn of data y(,1), y_kmeans ==0 means cluster 1
plt.scatter(X[y_kmeans ==0, 0], X[y_kmeans ==0,1], s = 100, c='red',label='cluster1')
plt.scatter(X[y_kmeans ==1, 0], X[y_kmeans ==1,1], s = 100, c='blue',label='cluster2')
plt.scatter(X[y_kmeans ==2, 0], X[y_kmeans ==2,1], s = 100, c='green',label='cluster3')
plt.scatter(X[y_kmeans ==3, 0], X[y_kmeans ==3,1], s = 100, c='cyan',label='cluster4')
plt.scatter(X[y_kmeans ==4, 0], X[y_kmeans ==4,1], s = 100, c='magenta',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300,c='yellow',label='Centriods')
plt.title('Clusters of clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending score')
plt.legend()
plt.show()
```
